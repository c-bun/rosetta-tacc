```python
import sys
  
# setting path
sys.path.append('../src')

from AnalysisTools import *
```

```python
data = read_decoy_dir("../data/SBit104/", ".sc")
```

```python

def add_labels(data):
    data['position'] = data['decoy'].apply(lambda x: int(x.split('_')[0][1:]))
    data['mutcode'] = data['decoy'].apply(lambda x: x.split('_')[0][1:]+x.split('_')[1][:3])
    data['AA'] = data['decoy'].apply(lambda x: x.split('_')[1][:3])
    return data

add_labels(data)
data.sample(10)
```

```python
datat = get_top(data, 'mutcode', 'I_sc', 1)
datat.sample(10)
```

```python
import plotnine as pn

(
    pn.ggplot(datat, pn.aes('position', 'I_sc', color='AA'))
    + pn.geom_point()
)
```

```python
pn.options.figure_size = (20, 6)

(
    pn.ggplot(datat[(datat['position']>148) & (datat['position']<156)], pn.aes('AA', 'I_sc'))
    + pn.geom_point()
    + pn.facet_wrap('position')
)
```

```python
AA_CLASS = {
    'GLY':'nonpolar',
    'ALA':'nonpolar',
    'VAL':'nonpolar',
    'LEU':'nonpolar',
    'ILE':'nonpolar',
    'MET':'nonpolar',
    'PRO':'nonpolar',
    'PHE':'nonpolar',
    'TRP':'nonpolar',
    'SER':'polar',
    'THR':'polar',
    'TYR':'polar',
    'CYS':'polar',
    'ASN':'polar',
    'GLN':'polar',
    'ASP':'acidic',
    'GLU':'acidic',
    'LYS':'basic',
    'HIS':'basic',
    'ARG':'basic',
}

def assign_class(data):
    data['aa class'] = data['AA'].apply(lambda x: AA_CLASS[x])
    return data

assign_class(datat)
datat.sample(5)
```

```python
(
    pn.ggplot(datat, pn.aes('position', 'I_sc', color='aa class'))
    + pn.geom_point()
)
```
