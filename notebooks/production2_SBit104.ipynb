```python
import sys
  
# setting path
sys.path.append('../src')

from AnalysisTools import *
```

```python
data = read_decoy_dir("../data/SBit104/", ".sc")
```

```python

def add_labels(data):
    data['position'] = data['decoy'].apply(lambda x: int(x.split('_')[0][1:]))
    data['mutcode'] = data['decoy'].apply(lambda x: x.split('_')[0][1:]+x.split('_')[1][:3])
    data['AA'] = data['decoy'].apply(lambda x: x.split('_')[1][:3])
    return data

add_labels(data)
data.sample(10)
```

```python
datat = get_top(data, 'mutcode', 'I_sc', 1)
datat.sample(10)
```

```python
import plotnine as pn

(
    pn.ggplot(datat, pn.aes('position', 'I_sc', color='AA'))
    + pn.geom_point()
)
```

```python
pn.options.figure_size = (20, 6)

(
    pn.ggplot(datat[(datat['position']>148) & (datat['position']<156)], pn.aes('AA', 'I_sc'))
    + pn.geom_point()
    + pn.facet_wrap('position')
)
```
